```python
n18 = sensor18.set_index("id")
```

```python
temp_y18 = pd.concat([temp18,n18],axis = 1)
```

```python
temp_y18.corr()
```

```python
plt.figure(figsize=(5,5))
ax = sns.heatmap(data = temp_y18.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Blues')

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.show()
```
